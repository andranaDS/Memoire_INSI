```python
import numpy as np
import pandas as pd
import sys
import matplotlib.pyplot as plt
import seaborn as sns

import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('energy_antananarivo_cleaned.csv', index_col=0, parse_dates=True)
```

Dans la suite de notre tude, on va diviser l'apprentissage en 2 partis:
   - un modle pour le nombre de demande en lectricit
   - un modle pour l'energie requise

```python
# Commencons par la modlisation du nombre de demande en lectricit
energie_df = df.drop('tot_energy', axis=1)
```

**Etape 1**: partition du dataset en set d'entrainement et set de test.  

Pour le set d'entrainement, on va utiliser 90 pourcente du dataset et le reste pour le set de test

```python
from sklearn.model_selection import train_test_split
```

```python
# Crer des ensembles de train et de test
train_df, test_df = train_test_split(energie_df, test_size=0.1, shuffle=False)

# Afficher les tailles des ensembles
print("Taille de l'ensemble d'entranement :", len(train_df))
print("Taille de l'ensemble de test :", len(test_df))
```

```python
train_df
```

On voit que le set d'entrainement par de l'index **2000-01-01**  l'index **2021-07-01** avec une longueur de **259**.  

L'etape suivante est de mettre en echelle les donnes

**Etape 2**: mise en echelle des donnes

Comme nous alimentons les rseaux neuronnes par avec les sorties en retour (label ) donc les labels  doivent tre mise  l'echelle.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
# Mise  l'echelle du set d'entrainement
scaler.fit(train_df)
```

```python
# Transformation de l'ensemble de donnes
scaled_train = scaler.transform(train_df)
scaled_test = scaler.transform(test_df)
```

**Etape 3**: cration de batch ou lot d'entrainement


```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
lenght = 12
batch_size = 1

# x: c'est l'index (date_dep_dmd)
# y: c'est la valeur predit (tot_nb_dmd)
# Et scaled_train contient ces 2 variables, d'o
generator = TimeseriesGenerator(scaled_train,
                                scaled_train,
                                length=lenght,
                                batch_size=batch_size)
```

```python
len(scaled_train)
```

```python
len(generator)
```

**X** correspond au 12 premiers valeurs. Il y a **12** valeur dans X car c'est le paramtre **length** qu'on avait defini un peu plus haut. 
Donc X va essayer de predir la troisime valeur qui est **y**.

### Important:  
**lentgh** ne doit pas tre choisi au hasard; il doit tre une valeur suffisante pour prendre en compte une **saisonnalit complete** ou autre effet de rptition.  
Et dans notre cas, une saisonnalit complte equivaut  12 mois. donc **length=12**

**Etape 4**: Cration du modle

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM
```

```python
n_features = 1 # c'est le nombre de features. ici il n'y a qu'un qui est 'tot_nb_dmd'
```

```python
model = Sequential()

# Ajout d'une couche
model.add(SimpleRNN(50, input_shape=(lenght,n_features)))
model.add(Dense(1))

# Etant donne qu'on veuille predir une valeur continue, donc on utilise l'optimiseur 'adam' et 'mse' comme 
# la fonction loss
model.compile(optimizer='adam', loss='mse')
```

```python
# Voyons voir le rsum du modle
model.summary()
```

**Etape 5**: Entrainement du modle

```python
model.fit(generator, epochs=5)
```

```python
# Evaluation des pertes
losses = pd.DataFrame(model.history.history)
losses.plot()
```

On constate une diminution sur les premiers epochs puis se stabilise et augmente lgrement
